```python
import pandas as pd
bank_data = pd.read_pickle('all_banks.pickle')
bank_data
```

```python
from pandas_profiling import ProfileReport
```

```python
# Flatten the Multi-Index
if isinstance(bank_data.columns, pd.core.indexes.multi.MultiIndex):
    bank_data.columns = ['.'.join(c) for c in bank_data.columns]
bank_data.columns = bank_data.columns.astype("str")
bank_data.head()
```

```python
bank_profile = ProfileReport(bank_data, title="The Banks' Data Exploratory Report", explorative=True)
bank_profile
```
